```python
import sys
sys.path.append("../../")

from environments.Key_To_Door.env import key_door_env

import numpy as np
```

```python
map_ascii_path='../../environments/Key_To_Door/env_configs/key_door_test_map_files/key_door_sample_map_mini_test.txt'
map_yaml_path = '../../environments/Key_To_Door/env_configs/key_door_test_map_files/key_door_sample_map_mini_test.yaml'

episode_timeout = 120
SEED = 121

env = key_door_env.KeyDoorEnv(
    map_ascii_path=map_ascii_path,
    map_yaml_path=map_yaml_path,
    representation="agent_position",
    field_x=5,
    field_y=5,
    episode_timeout=episode_timeout,
    seed=SEED
)

# * SETUP ROOMS BORDERS

env.ROOM_2_X1 = 7
env.ROOM_2_X2 = 11
env.ROOM_2_Y1 = env.ROOM_3_Y1 = 1
env.ROOM_2_Y2 = env.ROOM_3_Y2 = 5

env.ROOM_3_X1 = 13
env.ROOM_3_X2 = 17

env.PERCENT = 0.33 # * percent of total time T in the 1st and 3rd rooms (in the 2nd: T x (1-2*percent))



obs = env.reset()

done = False
timer = 0
obss = []

while not done:
    obss.append(obs)

    act = np.random.choice(env.action_space)
    obs, reward, done, info = env.step(act)

    timer += 1

print(timer)
obss.append(obs)

history = env._train_episode_history
env.close()
```

```python
resized_frames = []

# Iterate over each element in the frames list
for frame in history:
    # Determine the current size of the frame
    frame = np.sum(frame, axis=2)
    rows = len(frame)
    columns = len(frame[0])

    # Create a new array or image with the desired size
    resized_frame = np.zeros((rows * 50, columns * 50))

    # Copy the values from the original frame to the resized frame
    for i in range(rows):
        for j in range(columns):
            resized_frame[i*50:i*50+50, j*50:j*50+50] = np.ones((50,50)) * frame[i][j][np.newaxis, np.newaxis]

    # Append the resized frame to the new list
    resized_frames.append(resized_frame)

    # Append the resized frame to the new list
    resized_frames.append(np.array(resized_frame))
    
```

```python
from moviepy.editor import ImageSequenceClip
from moviepy.video.fx.resize import resize
import numpy as np

desired_resolution = (7*20,19*20)
original_aspect_ratio = 7 / 19
width = int(desired_resolution[0] * original_aspect_ratio)
height = desired_resolution[1]

observations = [np.repeat(o[:, :, np.newaxis], 3, axis=2) for o in resized_frames]

# observations = [o[:, :, 0] * 255 for o in observations]
mult = 255
for o in observations:
    o[:,:,0] *= mult/2
    o[:,:,1] *= mult/3
    o[:,:,2] *= mult/0.5

# Create ImageSequenceClip
clip = ImageSequenceClip(observations, fps=24)
# clip = clip.resize(width=width, height=height)
clip = resize(clip, width=width, height=height)

# Display the modified clip
clip.ipython_display(maxduration=120)
```
