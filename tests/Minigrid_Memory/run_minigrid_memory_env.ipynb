```python
env_name = {'type': 'Minigrid', 'name': 'MiniGrid-MemoryS13-v0'}
LENGTH = 31 # grid size

SEED = None
create_video = True

env = create_env(env_name, LENGTH, render=False) #Minigrid('MiniGrid-MemoryS13Random-v0', 31) 
obs = env.reset(seed=SEED)
plt.close()

done = False
frames = []

while not done:

    if create_video:
        rofl = env.render()
        plt.close()
        frames.append(rofl)

    act = env.action_space.sample()
    obs, reward, done, info = env.step(act)
    print(f'reward: {reward}, done: {done}, info: {info}')

if create_video:
    rofl = env.render()
    plt.close()
    frames.append(rofl)

env.close()
```
