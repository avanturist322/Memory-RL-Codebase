## RepeatPrevious

```python
import popgym 


env = POPGymWrapper('RepeatPreviousMedium')
init_obs = env.reset(seed = 42)
print(f'Init observaion {init_obs}')
history = [init_obs[-1]]
# print(env.deck.__dict__)
print(f'Max episode lenght: {env.max_episode_steps}')
for i in range(env.max_episode_steps):
    if i >= env.max_episode_steps //2:
        action = history.pop()
    else:
        action = env.action_space.sample()
    obs, reward, terminated, info = env.step(action)
    if i < env.max_episode_steps //2:
        history.append(obs[-1])
    if i >= env.max_episode_steps:
        history.append(obs[-1])
    print(f'Step {i}, Action :{action}, obs: {obs}, reward: {reward}, terminated: {terminated}, info: {info}')
```
